```python
full_data.fillna(method='ffill', inplace=True)
```

```python
full_data.drop(columns = ['customer name', 'customer e-mail', 'country'], inplace = True)
full_data.head(3)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
full_data_scaled = pd.DataFrame(scaler.fit_transform(full_data), columns=full_data.columns)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(full_data_scaled.drop('car purchase amount', axis=1), full_data_scaled['car purchase amount'], test_size=0.2, random_state=42)
```

# model development , training and testing 

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[len(X_train.keys())]),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
])
```

```python
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.RMSprop())
```

```python
history = model.fit(X_train, y_train, epochs=50, validation_split=0.2, verbose=1)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

R2 = r2_score(y_test, y_pred)
print("R2 Score=",R2 )
```
